```python
df.sort_values(by=['date'], inplace=False, ascending=False)

df.head()
```

```python
Apple_Stock = df.loc[df['symbol'] == 'AAPL', ['high', 'date','volume','year']]

print(Apple_Stock)
```

```python

df['year'] = df['year'].astype('int64') 
df['month'] = df['month'].astype('int64') 
df['day'] = df['day'].astype('int64')
df[["year", "month", "day"]] = df["date"].str.split("-", expand = True)

df.head()
```

```python
# Apple Stock 1999 - 2004
Apple_Stock = df.loc[df['symbol'] == 'AAPL', ['high', 'date','volume','year']]

Tech_Bubble =  Apple_Stock.loc[Apple_Stock['year'] == '1999', ['high', 'date','volume','year']]
print(Tech_Bubble)
```

```python
plt.scatter(x=Tech_Bubble['volume'],y=Tech_Bubble['high'])

plt.title('High Stock Price vs Volume traded')
plt.xlabel('Volume')
plt.ylabel('High')

plt.show
```

```python
correlation_matrix = Tech_Bubble.corr(method='pearson')
    
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Apple Stock Price 1999')
plt.xlabel('Apple Stock Features')
plt.ylabel('Apple Stock Features')

plt.show()
```
